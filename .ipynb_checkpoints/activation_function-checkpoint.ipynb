```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 시그모이드 함수
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([-1.0, 1.0, 2.0])
y = sigmoid(x)

print(y)
```

```python
x = np.arange(-5.0, 5.0, 1.0)
x
```

```python
y = sigmoid(x)
y
```

```python
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# 소프트맥스
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    
    return y

x = np.arange(-5.0, 5.0, 1.0)
y = softmax(x)

plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# 오버플로로 인해 계산 안됨
a = np.array( [1010, 1000, 990] )
print( np.exp(a) / np.sum(np.exp(a)) )
```

```python
c = np.max(a) # c는 입력의 최댓값을 이용합니다.
print( a - c )
```

```python
# 해결 함수
def softmax(a):
    c = np.max(a)
    exp_a = np.exp( a - c ) # 오버플로 대책
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    
    return y

x = np.arange(-10.0, 10.0, 1.0)
y = softmax(x)

plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
#ReLU 함수
def relu(x):
    return np.maximum(0, x)

x = np.arange(-10.0, 10.0, 1.0)
y = relu(x)

plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```
