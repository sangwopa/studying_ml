```python
# 신경망 학습 시, 매개변수는 주어진 매개변수에 대한 손실함수 변화도(Gradient)에 따라 조정됨
# 이러한 변화도 계산을 위해 pytorch에서는 torch.autograd라는 자동미분 엔진 내장
import torch
x = torch.ones(5) # 입력
y = torch.zeros(3) # 기대 출력
w = torch.randn(5, 3, requires_grad=True) # 가중치
b = torch.randn(3, requires_grad=True) # 편향
z = torch.matmul(x, w) + b
print(x)
print(y)
print(w)
print(b)
print(z)
```

```python
loss = torch.nn.functional.binary_cross_entropy_with_logits(z, y)
loss
```

```python
from IPython.display import Image

Image(r'C:\Users\user\Documents\GitHub\studying_ml\Image\autograd.png')
# 위 연산 그래프 도식화
# w,b 는 최적화를 해야하는 매개변수
# 이러한 변수들에 대한 손실함수의 변화도를 계산할 수 있어야 함
# 이를 위해 해당 tensor에 required_grad 속성 설정
```

```python
# grad_fn: 역방향 전파 함수에 대한 참조

print(f"Gradient function for z = {z.grad_fn}")
print(f"Gradient function for loss = {loss.grad_fn}")
```

```python
# 변화도 계산
loss.backward()
print(w.grad)
print(b.grad)
```

```python
# requires_grad=True인 모든 tensor들은 연산 기록을 추적하고 변화도 계산을 지원
z = torch.matmul(x, w) + b
print(z.requires_grad)
```

```python
# 순전파 연산만 필요한 경우 torch.no_grad()를 통해 추적 중지 가능
with torch.no_grad():
    z = torch.matmul(x, w)
print(z.requires_grad)
    
```

```python
# detach를 통해 동일하게 추적 중지 가능
z = torch.matmul(x, w) + b
z_det = z.detach()
print(z_det.requires_grad)
```

```python
# 변화도 추적을 멈춰야 하는 이유
# 사전 학습된 신경망을 미세조정(Fine tuning) 할 때, 신경망의 일부 매개변수를 고정된 매개변수로 표시
# 변화도 추적을 하지 않는 연산이 더 효율적
```
